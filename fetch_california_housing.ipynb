#Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
```

#Load Data

```python
data=fetch_california_housing()
```

```python
feature_names = data.feature_names
df=pd.DataFrame(data.data,columns=feature_names)

# Add the target column to the data
df['MedHouse'] = data.target
```

#Exploratory Data Analysis (EDA)

```python
df.sample(3)
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='HouseAge',y='MedHouse',data=df)
plt.title('HouseAge vs MedHouse')
plt.xlabel('HouseAge')
plt.ylabel('MedHouse')
plt.show()
#
```

```python
plt.figure()
sns.histplot(df['MedHouse'],label='MedHouse',bins=15,color='b',kde=True)
plt.legend()
plt.title('MedHouse Distribution')
plt.xlabel('MedHouse')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['Population'],df['MedHouse'])
plt.title('population vs MedHouse')
plt.xlabel('Population')
plt.ylabel('MedHouse')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Longitude',y='Latitude',data=df,hue='MedHouse')
plt.title('Longitude vs Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['MedHouse'], cmap='viridis')
plt.colorbar(label='Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('California Housing Prices by Location')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='MedHouse',data=df)
plt.title('MedHouse Distribution')
plt.xlabel('MedHouse')
plt.ylabel('Frequency')
plt.show()
```

```python
df.head()
```

```python
def age_group(age):
  if age<=10:
    return 'New_Homes'
  elif age<=30:
    return 'mid_age'
  elif age<=50:
    return 'old_homes'
  else:
    return 'very_old_homes'
```

```python
df['Home_Age_cat']=df['HouseAge'].apply(age_group)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Home_Age_cat',data=df)
plt.title('Home_Age_cat Distribution')
plt.xlabel('Home_Age_cat')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
sns.barplot(x='Home_Age_cat',y='MedHouse',data=df)
plt.title('Home_Age_cat vs MedHouse')
plt.xlabel('Home_Age_cat')
plt.ylabel('MedHouse')
plt.show()
```



*   Most houses are between 11 and 30 years old

*  the most expensive houses are over 50 years old
